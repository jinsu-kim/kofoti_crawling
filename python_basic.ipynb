# 자료형
- 자료형이란 프로그래밍을 할 때 쓰이는 숫자, 문자열 등 자료 형태로 사용하는 모든 것을 뜻합니다. <br>
  **프로그램의 기본이자 핵심 단위가 바로 자료형입니다.**
- 계산 프로그램을 만들려면 어떤 것을 계산할지부터 알아야 하고, 데이터베이스 프로그램을 만들려면 어떤 자료를 저장할지부터 알아야 하는 것처럼 <br>
  **입력 형태가 어떻게 되는지 정확히 알아야 합니다.**

## 1. 숫자형

## 1-1. 정수형

````python
# 양의 정수, 음의 정수, 0이 있습니다.

number_1 = 123
number_2 = -178
number_3 = 0
````

````python
# 변수의 자료형을 확인하기 위해서는 type(변수명)을 이용하시면 됩니다.
# 'int' 는 정수라는 의미로 integer 의 줄임말입니다.

print( type(number_1) )
print( type(number_2) )
print( type(number_3) )
````

## 1-2. 실수형

````python
# 정수형과 마찬가지로 양의 실수, 음의 실수, 0.0이 있습니다.
# 'float' 은 floating point (부동 소수점)의 줄임말입니다.

number_4 = 1.2355
number_5 = -23.562
number_6 = 0.0

print( type(number_4) )
print( type(number_5) )
print( type(number_6) )
````

## 1-3. 연산자

````python
# 사칙연산 또는 논리연산을 할 수 있습니다.

a = 3
b = 4

print( a+b )
print( a*b )
print( a**b )
print( b/a )
print( b//a )
print( b%a )
````

````python
# 논리 연산
# True 또는 False를 반환합니다.

print( a>b )
print( a<b )
print( a==b )
print( a>=b )
print( a<=b )
````

## 2. 문자열

````python
# 문자열은 따옴표('') 또는 쌍따옴표("") 안에 텍스를 입력하여 정의합니다.
# 아래 11번 라인 결과물로 ''으로 문자열을 정의한 것과 ""으로 문자열을 정의한 것이 같다는 것을 알 수 있습니다.
# == 은 논리 연산자로, == 기준으로 좌, 우의 데이터가 같은지를 확인할 수 있습니다.
# 같으면 True, 다르면 False를 반환합니다.

string_1 = "hello world"
string_2 = 'hello world'

print( type(string_1) )
print( type(string_2) )
print( string_1 == string_2 )
````

````python
# 따옴표로 묶지 않으면 에러가 발생합니다.
# invalid syntax: 문법에 맞지 않다는 얘기입니다.

string_3 = hello world
````

````python
# 여러 줄에 걸쳐서 문자열을 정의할 수 있습니다.
# """ """ 안에 텍스트를 입력하면 됩니다.

string_5 = """
I have a cat.
I am hungry.
I live in Seoul.
"""

print(string_5)
````

## 3. 리스트
**리스트는 배열의 한 종류로 여러 데이터를 묶어서 저장하는 자료형입니다.**

### 3-1. 리스트

````python
# 대괄호 안에 요소들을 나열하거나 [ 요소1, 요소2, 요소3 ]
# 내장 함수 list()를 이용하여 정의할 수 있습니다. list( (요소1, 요소2, 요소3) )

list_1 = ['dog', 'cat', 'pig', 'horse', 'sheep']
list_2 = list( ('dog', 'cat', 'pig', 'horse', 'sheep') )

print( list_1 )
print( type(list_1) )

print( list_2 )
print( type(list_2) )

print( list_1 == list_2 )
````

### 3-2. 인덱싱

````python
# 리스트에서 나열 순서를 이용해 특정 요소를 가져올 수 있습니다.
# 파이썬에서 인덱스는 0부터 시작합니다.

print( list_1[0] )
print( list_1[1] )
print( list_1[4] )

print( list_1[-1] )  # 마지막 인덱스를 알 수 없는 경우 -1을 입력하면 마지막 요소를 가져올 수 있습니다.
````

````python
# 요소의 개수가 5개이기 때문에 5를 입력하면 에러를 발생시킵니다.
# IndexError: list index out of range: 인덱스가 범위를 넘어섰다는 메세지입니다.

print( list_1[5] )
````

### 3-3. 슬라이싱

````python
# 리스트의 전체 또는 일부를 가져올 수 있습니다.
# colons (:) 을 기준으로 좌측은 시작하는 인덱스, 우측은 마지막 인덱스를 입력합니다.
# 주의해야할 것은 마지막 인덱스를 입력했지만 실제로 가져오는 것은 마지막 인덱스 앞의 요소까지입니다.

print( list_1[3] )
print( list_1[4] )
print( list_1[-1] )

print( list_1[0:3] )  # horse를 가져오지 않음
print( list_1[1:4] )  # sheep를 가져오지 않음
print( list_1[2:-1] )  # sheep를 가져오지 않음
````

### 3-4. 리스트의 요소

````python
# 리스트에는 문자열, 배열, 숫자 모두 포함될 수 있습니다.
# 리스트의 요소는 자료형에 구애받지 않습니다.

list_3 = [1, 2, 3, 'dog', 'cat', '철수', '학교', [190, 178, 184]]

print( type(list_3[0]) )
print( type(list_3[4]) )
print( type(list_3[6]) )
print( type(list_3[-1]) )
````

### 3-5. 리스트의 요소 추가

````python
# 리스트는 .append라는 함수로 요소를 추가할 수 있습니다.

list_3.append('추가_1')
print( list_3 )
list_3.append('추가_2')
print( list_3 )
list_3.append('추가_3')
print( list_3 )
````

## 4. 딕셔너리
**딕셔너리는 키워드와 값을 매핑시켜서 저장하는 자료형입니다.**

## 4-1. 딕셔너리

````python
# 딕셔너리의 기본 구조는 중괄호 안에 key와 value를 짝짓는 것입니다.
# { key: value }
# 일반적으로 key는 문자열, 정수형 데이터를 쓰고
# value에는 문자열, 정수형, 배열을 씁니다.

dict_1 = { 'key1':'val1', 'key2':'val2', 'key3':'val3', 'key4':'val4' }
dict_2 = { 'num1': 24, 'num2': 53, 'num3': 87 }
dict_3 = { 1:23, 2:56, 3:98 }
dict_4 = { '학생': ['철수','영희','시현'], '키': [173, 163, 178], '거주지':['서울','수원','의정부'] }

print( dict_1 )
print( dict_2 )
print( dict_3 )
print( dict_4 )
````

````python
# 잘못된 예
# 'key1'과 달리 key1은 변수로 인식되는데, key1이 선언된 적이 없기때문에 데이터가 없어서 에러를 발생시킵니다.
# 나머지도 변수들도 마찬가지입니다.
# NameError: name 'key1' is not defined: key1이 선언된 적이 없다는 메세지입니다.

dict_5 = {key1: value1, key2: value2, key3: value3}
````

# 제어문
- 제어문은 특정 조건에서 프로그램의 동작을 정의합니다.
- 조건문, 반목문이 있습니다.

## 1. 조건문

### 1-1. 조건문의 기본 구조

````python
# if 뒤의 조건식이 True (참) 이면 그 아래 구문이 실행됩니다.
# if 뒤 조건식의 끝은 반드시 colons(:)으로 마무리 해줍니다.
# if 뒤의 조건식이 False (거짓) 이면 그 아래 구문이 실행되지 않습니다.

if True:
    print('실행합니다.')

if False:
    print('실행하지 않습니다.')
````

````python
# 5가 3보다 크기때문에 아래 print()가 실행됩니다.

if 3 < 5:
    print('참입니다.')
````

````python
# 5가 3보다 크지 않기때문에 아래 print()가 실행되지 않습니다.

if 3 > 5:
    print('참입니다.')
````

````python
# 1은 [1,2,3,4,5,6,7,8] 안에 포함되기 때문에 참이고, 아래 구문이 실행됩니다.

list_4 = [1,2,3,4,5,6,7,8]

if 1 in list_4:
    print('1이 있네요')
````

## 2. 반복문

### 2-1. for 문의 기본 구조

````python
# for 뒤에 오는 배열이 있을 때 그 요소를 하나씩 가져오고, 마지막 요소를 가져올 때까지 반복합니다.
# 반복 조건식이 끝나는 부분은 반드시 colons(:)으로 마무리 해줍니다.
# range() 함수는 반복문을 작성할 때 많이 사용합니다.
# range(5)에서 숫자를 하니씩 가져와서 num 이라는 변수 안에 넣고, 그것을 출력했습니다.
# range()는 기본적으로 0부터 시작합니다.

for num in range(5):
    print( num )
````

## 3. 조건문과 반복문 조합해서 사용하기

````python
# 1에서 10까지의 숫자를 5를 기준으로 다른 동작을 실행해보겠습니다.
# range() 함수에 시작 숫자, 끝 숫자를 지정할 수 있습니다.

for num_ in range(1,11):  # 1~10
    if num_ < 5:
        print(num_, '5보다 작습니다.')
    elif num_ > 5:
        print(num_, '5보다 큽니다.')
    else:
        print('5입니다.')
````

````python
# 더 큰 숫자를 이용하는 코드를 작성해보겠습니다.

score_list = [82, 76, 28, 48, 96, 100, 67, 70]

for score in score_list:
    if score >= 80:
        print(score, '훌륭합니다')

    elif score < 80 and score >= 60:
        print(score, '분발하세요')

    else:
        print(score, '열심히 하세요')
````

# 정규표현식
- 크롤링을 하면 긴 문자열 중 우리에게 필요한 텍스트만 가져와야하는 경우가 있습니다.
- 이 때 정규표현식을 사용합니다.
- re: Regualar Experession

## 1. re.search()

````python
# import는 필요한 라이브러리를 불러오는 기능입니다.
# re 라는 라이브러리는 정규표현식을 사용할 수 있는 함수들을 모아놓은 것입니다.
# re 다음 dot(.)을 찍고 search라는 함수를 호출했습니다.
# 쓰임은 괄호 안에 쉼표를 기준으로 앞은 패턴, 뒤는 탐색하고자하는 문자열입니다.
# ca.e는 패턴, cafe는 탐색 대상이 됩니다.
# search는 탐색 대상 앞에서부터 스캔하며 패턴이 있는지 확인합니다.
# ca.e에서 (.)은 모든 문자를 대체할 수 있는 특별한 규칙이 있습니다.
# (.)을 기준으로 앞에는 ca가 있고 뒤에는 e가 있는 모든 문자를 가져올 수 있습니다.

import re   # regular expression

print( re.search('ca.e', 'cafe') )
print( re.search('ca.e', 'ca*e') )
print( re.search('ca.e', 'ca2e') )
````

````python
# search 함수는 앞에서부터 탐색하는데, 패턴에 부합하는 텍스트 중 가장 앞에 있는 것을 가져옵니다.
# 'ca.e' 패턴에 부합하는 문자열은 'cafe', 'case'가 있지만 'cafe'만 출력됩니다.

print( re.search('ca.e', 'cafecase') )
````

## 2. re.findall()

````python
# findall 함수는 앞에서부터 탐색하지만, 패턴에 부합하는 텍스트를 모두 가져옵니다.
# 패턴에 부합하는 텍스트를 모두 가져와서 리스트로 만들어줍니다.
# search는 출력 메세지가 어려우니 앞으로는 findall만 쓰겠습니다.

print( re.findall('ca.e', 'cafecase') )
````

## 3. + 그리고 *

````python
# + 그리고 * 을 패턴에 사용할 때는 이 특수문자의 바로 앞의 문자(아래 예시에서는 a)가 중요합니다.
# c, e 사이에 등장하는 a가 여러번 등장하는 텍스트를 가져올 때 패턴에 + 또는 *를 사용합니다.
# 차이는 +는 a가 한 개 이상, *는 a가 0개 이상 등장하는 텍스트를 가져옵니다.

print( re.findall('ca*e', 'ce') )
print( re.findall('ca*e', 'caaaaaaae') )

print( re.findall('ca+e', 'ce') )
print( re.findall('ca+e', 'caaaaaaae') )
````

## 4. \ d {  }

````python
# \d{} 패턴은 문자열에서 숫자를 가져올 때 사용합니다.
# {}에는 연이어 등장하는 숫자의 개수를 명시합니다.
# \d{3} 이라면 연이어 등장하는 숫자가 3개인 경우 텍스트를 가져옵니다.
# \d{1,3} 이라면 연이어 등장하는 숫자가 1개에서 3개인 경우 가져옵니다.

print( re.findall('\d{3}', '개 24마리, 닭 128마리, 병아리 4327마리, 소 4마리') )
print( re.findall('\d{1,3}', '개 24마리, 닭 128마리, 병아리 4327마리, 소 4마리') )
````

## 5. 가격 텍스트 가져오기

````python
# 웹 사이트에서 일반적으로 가격은 000,000원 이런 식으로 표시됩니다.
# comma(,)를 기준으로 좌측은 1-3개의 숫자가 등장할 수 있고, 우측은 무조건 3개 숫자입니다.
# comma(,)는 dot(.)처럼 특별한 규칙이 없기때문에 텍스트 그대로 취급합니다.

print( re.findall('\d{1,3},\d{3}', '\n무신사 회원가X\n\n>\n 회원할인 43,700   비회원55,200원\n') )
````

## 6. [ ]

````python
# 웹 사이트에서 취급하는 항목이 10만 원 단위일 수도 있지만 때에 따라서 100만 원 단위일 수도 있습니다.
# 1만 단위, 10만 단위, 100만 단위를 모두 가져올 수 있는 패턴을 작성해보겠습니다.
# []는 안의 긴 패턴을 묶어서 하나의 문자처럼 취급합니다.

print( re.findall('[\d{1,3},]+\d{3}', '\n무신사 회원가X\n\n>\n 회원할인 982,700   비회원1,155,200원\n ') )
````

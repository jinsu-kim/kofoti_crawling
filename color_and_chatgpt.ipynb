# 색상정보 추출

```python
# 한 번만 실행시키시면 됩니다.
!pip install sklearn

from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
# 색상 추출 함수
# 함수가 무슨 역할을 하는지만 아시면 돼요.
# 이 셀에 있는 코드를 하나하나 확인하실 필요는 없습니다.

from collections import Counter   # 입력된 데이터에서 요소의 등장 횟수를 카운트

def show_img_compar(img_1, img_2 ):  # 이미지를 동시에 보여주기
    f, ax = plt.subplots(1, 2, figsize=(8,4))
    ax[0].imshow(img_1)
    ax[1].imshow(img_2)
    ax[0].axis('off')
    ax[1].axis('off')
    f.tight_layout()
    plt.show()

def palette_perc(k_cluster):
    width = 300
    palette = np.zeros((50, width, 3), np.uint8)

    n_pixels = len(k_cluster.labels_)
    counter = Counter(k_cluster.labels_) # count how many pixels per cluster
    perc = {}
    for i in counter:
        perc[i] = np.round(counter[i]/n_pixels, 2)
    perc = dict(sorted(perc.items()))

    result = {}
    ratio_dict = {i:np.round(counter[i] / n_pixels, 2) for i in counter}
    for idx, color in enumerate(k_cluster.cluster_centers_):
        ratio = ratio_dict[idx]
        rgb_color = '#%02x%02x%02x'.upper() % (int(color[0]),int(color[1]),int(color[2]))
        result[rgb_color] = ratio

    result_new = {}
    before_sort = sorted(result.items(), reverse=True, key=lambda item:item[1])
    for key, value in before_sort:
        result_new[key] = value

    step = 0
    for idx, centers in enumerate(k_cluster.cluster_centers_):
        palette[:, step:int(step + perc[idx]*width+1), :] = centers
        step += int(perc[idx]*width+1)

    return result_new, palette
```

```python
# 입력 이미지, 메인 색상과 비율 추출하기

clt = KMeans(n_clusters=5)


file_name = '/Users/jinsu/PycharmProjects/crawling/crawling_sele/16_Jacquemus_레터링 로고 크롭 가디건.jpg'
img_rgb = Image.open(file_name)   # 경로에 있는 파일을 읽어오기
img_grey = img_rgb.convert('L')   # grey scale로 변환


img_rgb_array = np.array(img_rgb)  # RGB 이미지를 배열로 만들기, array: 배열
img_grey_array = np.array(img_grey)  # Grey Scale 이미지를 배열로 만들기


# 옷 영역만 추출하기
mask = img_grey_array != 255
clt_input = img_rgb_array[mask]


# 색상 추출하기 위해 세팅
clt_img = clt.fit(clt_input)


# 입력 이미지, 색상 비율, 파레트 동시에 보여주기
color_text, color_palette = palette_perc(clt_img)
show_img_compar(img_rgb_array, color_palette)
print(color_text)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable

fig = plt.figure(figsize=(15,5))

ax1 = fig.add_subplot(1, 3, 1)
im1 = ax1.imshow(img_rgb_array)
ax1.set_title('img_rgb')


ax2 = fig.add_subplot(1, 3, 2)
ax2.imshow(img_grey_array, cmap='gray')
ax2.set_title('img_grey')


mask = np.where(img_grey_array<255, 1, 0)
ax3 = fig.add_subplot(1, 3, 3)
ax3.imshow(mask, cmap='gray')
ax3.set_title('mask')
```

# ChatGPT API 이용하기

## ChatGPT 사용시 유의하셔야하는 부분

- ChatGPT는 만능이 아닙니다.
- ChatGPT는 일반적인 딥러닝 학습 방식을 따르기때문에 학습한 데이터에 의해 대답의 질이 결정됩니다.
- 그렇기 때문에 아주 최신인 (최근 1,2달 이내) 정보에 대한 답은 할 수 없습니다.
- Chat GPT가 일반적으로 좋은 대답을 내놓지만 때로는 정확하지 않은 대답을 하는 경우도 있습니다.
- Cross check가 반드시 필요합니다 !!!!!
- 프롬프트(또는 질문)를 작성하실 때는 최대한 자세하게 써주셔야합니다.
> 바람직한 질문 예: <br>
    > when using selenium, I faced this message "You have sent too many requests in a given amount of time." <br>
    > what should I do?<br><br>
> 바람직하지 않은 질문 예: <br>
    > I faced this message "You have sent too many requests in a given amount of time."
- 위 문장의 차이는 에러를 맞딱뜨린 상황을 적어줬는지 아닌지입니다.
- 간단하게라도 세부 정보를 줌으로써 ChatGPT가 제대로된 대답을 할 수 있도록 할 수 있습니다.
- 코드를 작성하는 경우도 마찬가지입니다.
- 그럴듯한 코드를 작성해줄 수는 있으나, 실행결과과 이상하거나 코드가 동작하지 않을 수 있습니다.
